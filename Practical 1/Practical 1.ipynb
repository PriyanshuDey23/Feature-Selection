# Feature Selection - Dropping constant features 

# In this step we will be removing the features which have constant features which are actually not important for solving the problem statement

# If in our dataset we have 500-600 features, it may not perform well 

# Small Example 

```python

import pandas as pd
data=pd.DataFrame({"A":[1,2,4,1,2,4],
                  "B":[4,5,6,7,8,9],
                  "C":[0,0,0,0,0,0],
                  "D":[1,1,1,1,1,1]})
```

```python
data # (these are all the input feature)
```

```python
# We will try to remove the column "c" and "D" because they have constant feature
# we are going to use a library from sklearn to remove the column i.e. variance Threshold
# variance threshold - feature selector that removes all low variance features
# if all value are same then mean, variance etc. all will be same
# threshold(0) - It is going to check the features which are having zero variance
# In this we are going to remove zero variance features
#i.e. whichever has zero variance it is going to remove
# check variance formula
from sklearn.feature_selection import VarianceThreshold
var_thres=VarianceThreshold(threshold=0)
var_thres.fit(data) # fit on data
```

```python
var_thres.get_support()
# true- We donot have to remove this features(non-zero variance)
# false- we have to remove this features(zero variance)
```

```python
data.columns[var_thres.get_support()]
```

```python
# we are going to remove the constant column
constant_columns = [ column for column in data.columns
                   if column not in data.columns[var_thres.get_support()]]
# constant_column will store the zero columns
# If not present then store
```

```python
constant_columns
```

```python
print(len(constant_columns))
```

```python
for feature in constant_columns:
    print(feature)
```

```python
# drop the 2 column
data.drop(constant_columns,axis=1)
```

```python
from sklearn.model_selection import train_test_split
# separate dataset into train and test 
x_train, x_test, y_train, y_test = train_test_split(
    df.drop(labels=['TARGET'], axis=1),
    df['TARGET'],
    test_size=0.3,
    random_state=0)

x_train.shape, x_test.shape
# one value has got into dependent feature
```

# Lets apply the variance threshold

```python
from sklearn.feature_selection import VarianceThreshold
var_thres=VarianceThreshold(threshold=0)
var_thres.fit(x_train)
# fit on x train( output will be machine understandable)
# transform on x test ( convert into machine understand format)
# it is done on x test to avoid over fitting
```

```python
#finding non constant features
# total length 
sum(var_thres.get_support())
# 284 value which are true
# 86 are duplicate or constant features
```

```python
var_thres.get_support()
```

```python
constant_column = [ column for column in x_train.columns
                   if column not in x_train.columns[var_thres.get_support()]]
print(len(constant_column))
```

```python
constant_column
```

```python
for column in constant_column:
    print(column)
```

```python
# dropping of constant column 
x_train.drop(constant_column,axis=1)
```
